```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N= 50
dx = 1.0 / N
```

```python
x = np.linspace(0,1,N+1)
y = np.ones(len(x))
for i in range(0,N):
    y[i+1] = y[i] + dx * (y[i] - 2*x[i]/y[i])
y1 = np.sqrt(1+2*x)
```

```python
plt.plot(x,y,x,y1)
# plt.plot(x,y1,color='y')
plt.show()
# draw(x,y,y1)
```
